```python
import snscrape.modules.twitter as sntwitter
import pandas as pd

# Creating list to append tweet data to
tweets_list2 = []

# Using TwitterSearchScraper to scrape data and append tweets to list
for i,tweet in enumerate(sntwitter.TwitterSearchScraper('fluoride OR fluorides OR fluoridation since:2020-01-01 until:2020-06-30').get_items()):
    if i>10000:
        break

    tweets_list2.append([tweet.username, tweet.date, tweet.id, tweet.content])
    
# Creating a dataframe from the tweets list above
tweets_df2 = pd.DataFrame(tweets_list2, columns=['Username', 'Datetime', 'Tweet Id', 'Text'])
tweets_df2.to_csv('Tweets.csv')
tweets_df2
```
